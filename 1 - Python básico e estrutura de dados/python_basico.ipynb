```python
# Criando uma variável do tipo string
frase = "Python para análise de dados"
```

```python
# Fatiamento de string
frase[0]
```

```python
# Fatiamento de string em 10 posiçcões iniciando no index 0 
frase[0:10]
```

```python
# Fatiamento de string a partir da décima posição
frase[10:]
```

```python
# Fatiamento de string em 17 posições iniciando no index 0 
frase[:17]
```

```python
# count: permite a contagem de elementos dentro da variável
frase.count("dados")
```

```python
# lower: coverte os elementos de uma string para minúsculo.
frase.lower()
```

```python
# upper: coverte os elementos de uma string para maiúsculo.
frase.upper()
```

```python
# replace: permite substituir elementos de uma string.
frase.replace("Python", "Frases")
```
